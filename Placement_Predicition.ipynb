This is a Supervised Machine Learning, to predict whether a student will get placed or not based on the various factors, and built using avaiblae Dataset

The input:
gender,
sslc percentage,
sslc board,
puc board,
puc percentage,
puc stream,
degree percentage,
degree stream,
work experience,


output:
placed or not

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSNeLrhjSBnnKZw2-1Ji8mEmZ-UlAJIVdeUgXR7XuSSY1CcnwZBkxqTleYG73lWHBuh-55nAwc5OdU1/pub?gid=1617506194&single=true&output=csv')
data
```

```python
data.columns
data['status'].value_counts()

data[(data['degree_t']=='Sci&Tech') & (data['status']=='Placed')].sort_values(by="salary", ascending=False).head()
```

```python
# Now convert string to number, aka encoding
data['gender']=data['gender'].map({'M':1, 'F':0})
data['ssc_b']=data['ssc_b'].map({'Central':1, 'Others':0})
data['hsc_b']=data['hsc_b'].map({'Central':1, 'Others':0})
data['hsc_s']=data['hsc_s'].map({'Science':2, 'Arts':1, 'Commerce':0})
data['degree_t']=data['degree_t'].map({'Sci&Tech':2, 'Comm&Mgmt':1, 'Others':0})
data['specialisation']=data['specialisation'].map({'Mkt&Fin':1, 'Mkt&HR':0})
data['workex']=data['workex'].map({'Yes':1, 'No':0})
data['status']=data['status'].map({'Placed':1, 'Not Placed':0})
data.head()
```

```python
# Removing Unneccessary Columns, here I am removing MBA also, coz I am a CSE Student
# sl_no, salary, mba_p
data = data.drop(['sl_no', 'salary', 'mba_p'], axis=1)
```

```python
X = data.drop(['status'], axis=1)
Y = data['status']

from sklearn.model_selection import train_test_split
X_train, Y_train, X_test, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
Y.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

m1 = LogisticRegression()
m1.fit(X_train, Y_train)

m2 = KNeighborsClassifier()
m2.fit(X_train, Y_train)

m3 = svm.SVC()
m3.fit(X_train, Y_train)

m4 = DecisionTreeClassifier()
m4.fit(X_train, Y_train)

m5 = RandomForestClassifier()
m5.fit(X_train, Y_train)

m6 = GradientBoostingClassifier()
m6.fit(X_train, Y_train)
```

```python
pred_by_m1 = m1.predict(X_test)
pred_by_m2 = m2.predict(X_test)
pred_by_m3 = m3.predict(X_test)
pred_by_m4 = m4.predict(X_test)
pred_by_m5 = m5.predict(X_test)
pred_by_m6 = m6.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
ms1 = accuracy_score(Y_test, pred_by_m1)
ms2 = accuracy_score(Y_test, pred_by_m2)
ms3 = accuracy_score(Y_test, pred_by_m3)
ms4 = accuracy_score(Y_test, pred_by_m4)
ms5 = accuracy_score(Y_test, pred_by_m5)
ms6 = accuracy_score(Y_test, pred_by_m6)

# models = [m1, m2, m3, m4, m5, m6]
# accuracy_scores = []
# for model in models:
#     y_pred = model.predict(X_test)
#     accuracy = accuracy_score(Y_test, y_pred)
#     accuracy_scores.append(accuracy)

# for i, accuracy in enumerate(accuracy_scores):
#     print(f"ms{i+1} = {accuracy}")
```
